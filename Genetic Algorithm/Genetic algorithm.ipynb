## Population class

```python
import numpy as np
class chromosome:
    def __init__(self, ch_length,genes=None):
        self.ch_length=ch_length
        self.genes = genes if genes else self.generate_genes()
        self.fitness = fitness_function(self.genes)

    def generate_genes(self):
        return [np.random.choice([0, 1]) for _ in range(self.ch_length)]
        

class Population:
    def __init__(self, size=10, chromosome_length=10):
        
        self.chromosomes = [chromosome(chromosome_length) for _ in range(size)]
        self.size = len(self.chromosomes)
        # self.chromosome_fitness
```

## Fitness function

```python
def fitness_function(x): # objective function #sphere

    fitness_value = np.sum(np.power(x,2))
    return fitness_value
```

## selection

```python
def selection(population):
    fitness = np.array([ch.fitness for ch in population.chromosomes])
    min_fitness = min(fitness)
    if min_fitness<0:
        fitness = fitness+abs(min_fitness)+1
    normalized_fitness = fitness/sum(fitness)
    index_sorted_normalized_fitness = np.argsort(-normalized_fitness)
    sorted_normalized_fitness = -np.sort(-normalized_fitness)
    size = len(sorted_normalized_fitness)

    cumsum = np.zeros(size)
    for i in range(size):
        cumsum[i]=sum(sorted_normalized_fitness[i:])


    R = np.random.rand()
    parent1=None
    print(R,cumsum)
    for k in range(1,len(cumsum)):
        if R > cumsum[k]:
            parent1 = population.chromosomes[index_sorted_normalized_fitness[k-1]]
            break
    if parent1 is None:
        parent1 = population.chromosomes[index_sorted_normalized_fitness[len(cumsum)-1]]
        
    ############################## this block will work error free only for chromosome count >=3 all the time #####
    ############################fix this later ####################################################################
    parent2 = parent1
    while parent2==parent1:
        R = np.random.rand()
        for k in range(1,len(cumsum)):
            if R > cumsum[k]:
                parent2 = population.chromosomes[index_sorted_normalized_fitness[k-1]]
                break
    ###############################################################################################################
            
    
    return [chromosome(len(parent1.genes),parent1.genes),chromosome(len(parent2.genes),parent2.genes)]
```

## crossover

```python
def crossover(parent1,parent2, Pc,crossoverName):
    if crossoverName=='double':
        gene_no = len(parent1.genes)
        ub = gene_no
        lb = 1
        cross_p1 = np.random.randint(lb,ub) #crossover point
        cross_p2=cross_p1
        while cross_p2==cross_p1:
            cross_p2 = np.random.randint(lb,ub)
        if cross_p1>cross_p2:
            cross_p1,cross_p2 = cross_p2,cross_p1
            
        part1 = parent1.genes[:cross_p1]
        part2 = parent2.genes[cross_p1:cross_p2]
        part3 = parent1.genes[cross_p2:]
        child1 = part1+part2+part3
        child1 = chromosome(len(child1),child1)
        
        
        part1 = parent2.genes[:cross_p1]
        part2 = parent1.genes[cross_p1:cross_p2]
        part3 = parent2.genes[cross_p2:]
        child2 = part1+part2+part3
        child2 = chromosome(len(child2),child2)
        
    elif crossoverName=='double':
        pass

    r1 = np.random.rand()
    if r1<Pc:
        child1 = child1
    else:
        child1 = parent1

    r2 = np.random.rand()
    if r2<Pc:
        child2 = child2
    else:
        child2=parent2

    return [child1,child2]
        
    
    
```

## mutation

```python
def mutation(child,Pm):
    gene_no = len(child.genes)
    for k in range(gene_no):
        r = np.random.rand()
        if r<Pm:
            child.genes[k]=np.logical_not(child.genes[k]).astype(int)
            child.fitness = fitness_function(child.genes)
            
    return child
```

## elitism(optional/problem based)

```python
def elitism(population, new_population, Er):
    
    m = len(new_population.chromosomes)
    
    elite_no = int(np.round(m*Er))
    print("elite_no:",elite_no)
    fitness = np.array([ch.fitness for ch in population.chromosomes])
    index_sorted = np.argsort(-fitness)

    for k in range(elite_no):
        new_population.chromosomes[k].genes = population.chromosomes[index_sorted[k]].genes
        new_population.chromosomes[k].fitness = population.chromosomes[index_sorted[k]].fitness

    return new_population
    
```
