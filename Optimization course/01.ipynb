### Design a table

Decision variables:
- length
- width

objective
- minimize weight= length * width

constraints:
subject to-
- 2<lenth<7
- 2<width<7


```python
# importing mplot3d toolkits, numpy and matplotlib
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

# syntax for 3-D projection
ax = plt.axes(projection ='3d')

# defining all 3 axis

x = np.linspace(2,7,51)
y = np.linspace(2,7,51)
z = x*y

# plotting
ax.plot3D(x, y, z, 'green')
plt.show()
```
